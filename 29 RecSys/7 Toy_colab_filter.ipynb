```python
def color_negative_red(val):
    color = 'red' if val < 0 else 'black'
    return 'color: %s' % color
```

```python
df_rec2= df_rec.style.applymap(color_negative_red).set_properties(**{
    'background-color': 'grey',
    'font-size': '10pt',
}).format(precision=3)
```
